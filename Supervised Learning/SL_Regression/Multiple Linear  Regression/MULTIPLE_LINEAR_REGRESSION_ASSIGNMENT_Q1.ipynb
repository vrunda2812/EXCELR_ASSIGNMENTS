# FOWARD SELECTION

```python
data = pd.read_csv("/content/50_Startups.csv")
```

```python
data
```

```python
data.columns = ['RDSpend', 'Admin', 'MarSpend', 'State', 'Profit']
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import statsmodels.formula.api as smf
model=smf.ols("Profit~RDSpend+Admin+MarSpend",data=data).fit()
```

```python
model
```

```python
## MODEL TESTING 
model.params
```

```python
# t-value// p-value
model.tvalues , np.round(model.pvalues,3)
```

```python
# THE R-SQ/ ADJ. R-SQ
model.rsquared , model.rsquared_adj
```

```python
## MODEL VALIDATION 
# (A) COLLINEARITY (for independent variables)

rq = smf.ols("RDSpend~ Admin + MarSpend",data=data).fit().rsquared
vif_rq = 1/(1-rq)

rq1 = smf.ols("Admin~ RDSpend + MarSpend",data=data).fit().rsquared
vif_rq1 = 1/(1-rq1)

rq2 = smf.ols("MarSpend~RDSpend + Admin",data=data).fit().rsquared
vif_rq2 = 1/(1-rq2)
```

```python
d1={'Variables':['RDS','ADMS','MKTS'],'Vif':[vif_rq,vif_rq1,vif_rq2]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# (B) RESIDUAL ANALYSIS

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'RDSpend',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'Admin',fig=fig)
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'MarSpend',fig=fig)
plt.show()
```

# MODEL DELETION DIAGNOSTICS

```python
(c,_)=model.get_influence().cooks_distance
c
```

```python
# VISUALIZATIONS FOR COOK'S DISTANCE 
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(c) , np.max(c)
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
## IMPROVING THE MODEL 
imp1 = data.drop(data.index[[49]],axis=0).reset_index(drop=True)
imp1
```

```python
while np.max(c)>0.5 :
    model=smf.ols("Profit~RDSpend + Admin + MarSpend",data=imp1).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    imp1=imp1.drop(imp1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    imp1
else:
    final_model=smf.ols("Profit~RDSpend + Admin + MarSpend",data=imp1).fit()
    final_model.rsquared , final_model.aic
    print("The model accuracy is improved to",final_model.rsquared)
```
