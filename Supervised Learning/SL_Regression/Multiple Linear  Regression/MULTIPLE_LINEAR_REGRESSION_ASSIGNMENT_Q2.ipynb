# FOWARD SELECTION

```python
data = pd.read_csv("/content/ToyotaCorolla.csv", encoding = 'latin1')
```

```python
data.columns
```

```python
data = pd.concat([data.iloc[:,2:4],data.iloc[:,6:7],data.iloc[:,8:9],data.iloc[:,12:14],data.iloc[:,15:18]],axis=1)
```

```python
data.columns
```

```python
data.columns = ['Price', 'Age', 'KM', 'HP', 'CC', 'Doors', 'Gears','QuartTax', 'Weight']
```

```python
data.head()
```

```python
# CORRELATION MATRIX
import seaborn as sns
sns.heatmap(data.corr(), annot=True)
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(data)
```

```python
import statsmodels.formula.api as smf
model=smf.ols("Price~Age+KM+HP+CC+Doors+Gears+QuartTax+Weight",data=data).fit()
```

```python
## MODEL TESTING 
model.params
```

```python
# t-value// p-value
model.tvalues , np.round(model.pvalues,3)
```

```python
# THE R-SQ/ ADJ. R-SQ
model.rsquared , model.rsquared_adj
```

```python
## MODEL VALIDATION 
# (A) COLLINEARITY (for independent variables)

rq = smf.ols("Age~KM+HP+CC+Doors+Gears+QuartTax+Weight",data=data).fit().rsquared
vif_rq = 1/(1-rq)

rq1 = smf.ols("KM~ Age+HP+CC+Doors+Gears+QuartTax+Weight",data=data).fit().rsquared
vif_rq1 = 1/(1-rq1)

rq2 = smf.ols("HP~KM+Age+CC+Doors+Gears+QuartTax+Weight",data=data).fit().rsquared
vif_rq2 = 1/(1-rq2)

rq3 = smf.ols("CC~KM+Age+HP+Doors+Gears+QuartTax+Weight",data=data).fit().rsquared
vif_rq3 = 1/(1-rq3)

rq4 = smf.ols("Doors~KM+Age+CC+HP+Gears+QuartTax+Weight",data=data).fit().rsquared
vif_rq4 = 1/(1-rq4)

rq5 = smf.ols("Gears~KM+Age+CC+Doors+HP+QuartTax+Weight",data=data).fit().rsquared
vif_rq5 = 1/(1-rq5)

rq6 = smf.ols("QuartTax~KM+Age+CC+Doors+Gears+HP+Weight",data=data).fit().rsquared
vif_rq6 = 1/(1-rq6)

rq7 = smf.ols("Weight~KM+Age+CC+Doors+Gears+QuartTax+HP",data=data).fit().rsquared
vif_rq7 = 1/(1-rq7)
```

```python
d1={'Variables':['Age','KM', 'HP', 'CC', 'Doors', 'Gears', 'QuartTax', 'Weight'],'Vif':[vif_rq,vif_rq1,vif_rq2,vif_rq3, vif_rq4, vif_rq5, vif_rq6, vif_rq7 ]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# (B) RESIDUAL ANALYSIS

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model,'KM',fig=fig)
plt.show()
```

```python
# MODEL DELETION DIAGNOSTICS
(c,_)=model.get_influence().cooks_distance
c
```

```python
#  VISUALIZATIONS FOR COOK'S DISTANCE 
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(c) , np.max(c)
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
## IMPROVING THE MODEL 
imp1 = data.drop(data.index[[80]],axis=0).reset_index(drop=True)
imp1
```

```python
while np.max(c)>0.5 :
    model=smf.ols("Price~Age+KM+HP+CC+Doors+Gears+QuartTax+Weight",data=imp1).fit()
    (c,_)=model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    imp1=imp1.drop(imp1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    imp1
else:
    final_model=smf.ols("Price~Age+KM+HP+CC+Doors+Gears+QuartTax+Weight",data=imp1).fit()
    final_model.rsquared , final_model.aic
    print("The model accuracy is improved to",final_model.rsquared)
```
