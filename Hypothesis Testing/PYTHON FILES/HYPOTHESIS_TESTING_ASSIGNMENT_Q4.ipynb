```python
error_free = pd.Series([271, 267, 269, 280])
```

```python
defective = pd.Series([29, 33, 31, 20])
```

```python
import scipy.stats as stats
stats.f_oneway(error_free, defective)
```

```python
pvalue = 1.437376025121072e-09
if pvalue > 0.05:
  print("H0 is rejected as the defective percentage for 4 centres is not equal")
else:
  print("H0 is accepted as the defective percentage for the 4 centres is equal")
```
